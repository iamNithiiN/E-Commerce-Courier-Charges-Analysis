```python
# Import Required Libraries
import pandas as pd
import numpy as np
```

```python
# Loading data
order_report  = pd.read_excel('Company X - Order Report.xlsx')
pincode_zones = pd.read_excel('Company X - Pincode Zones.xlsx')
SKU_Master    = pd.read_excel('Company X - SKU Master.xlsx')
cc_invoice    = pd.read_excel('Courier Company - Invoice.xlsx')
cc_rates      = pd.read_excel('Courier Company - Rates.xlsx')
```

```python
merged_data = pd.merge(order_report, cc_invoice, left_on='ExternOrderNo', right_on='Order ID')
```

```python
merged_data.head()
```

```python
merged_data = pd.merge(merged_data, pincode_zones, left_on='Warehouse Pincode', right_on='Warehouse Pincode')
```

```python
merged_data.head()
```

```python
merged_data = pd.merge(merged_data, pincode_zones, left_on='Customer Pincode_y', right_on='Customer Pincode')
```

```python
merged_data.head()
```

```python
merged_data = pd.merge(merged_data, SKU_Master, left_on='SKU', right_on='SKU')
```

```python
merged_data.head()
```

```python
for i in merged_data['Zone_y']:
    if i=="a":
        merged_data['Weight Slabs']=0.25
        merged_data['Forward Fixed Charge']=29.5
        merged_data['Forward Additional Weight Slab Charge']=merged_data['RTO Additional Weight Slab Charge']=23.6
        merged_data['RTO Fixed Charge']=13.6
    elif i=="b":
        merged_data['Weight Slabs']=0.50
        merged_data['Forward Fixed Charge']=33
        merged_data['Forward Additional Weight Slab Charge']=merged_data['RTO Additional Weight Slab Charge']=28.3
        merged_data['RTO Fixed Charge']=20.5
    elif i =="c":
        merged_data['Weight Slabs']=0.75
        merged_data['Forward Fixed Charge']=40.1
        merged_data['Forward Additional Weight Slab Charge']=merged_data['RTO Additional Weight Slab Charge']=28.9
        merged_data['RTO Fixed Charge']=31.9
    elif i=="d":
        merged_data['Weight Slabs']=1.25
        merged_data['Forward Fixed Charge']=45.4
        merged_data['Forward Additional Weight Slab Charge']=merged_data['RTO Additional Weight Slab Charge']=44.8
        merged_data['RTO Fixed Charge']=41.3
    elif i =="e":
        merged_data['Weight Slabs']=1.50
        merged_data['Forward Fixed Charge']=56.6
        merged_data['Forward Additional Weight Slab Charge']=merged_data['RTO Additional Weight Slab Charge']=55.5
        merged_data['RTO Fixed Charge']=50.7
```

```python
merged_data.head()
```

```python
merged_data.shape
```

```python
# Calculate total weight
merged_data['Total Weight'] = merged_data['Order Qty'] * (merged_data['Weight (g)'] / 1000)
```

```python
# Calculate weight slab as per X
merged_data['Weight Slabs'] = merged_data['Total Weight'].apply(lambda x: 0.25 if x <= 0.25 else 0.50 if x <= 0.50 else 0.75 if x <= 0.75 else 1.25 if x <= 1.25 else 1.50)
```

```python
merged_data
```

```python
# Calculate total weight as per Courier Company
# ... (use Charged Weight from cc_invoice)
merged_data['Total Weight (Courier Company)'] = merged_data['Charged Weight']
```

```python
# Calculate weight slab charged by Courier Company
merged_data['Weight Slabs (Courier Company)'] = merged_data['Total Weight (Courier Company)'].apply(lambda x: 0.25 if x <= 0.25 else 0.50 if x <= 0.50 else 0.75 if x <= 0.75 else 1.25 if x <= 1.25 else 1.50)
```

```python
# Calculate expected charge as per X
# ... (calculate based on weight slab, delivery zone, and COD charges)
merged_data['Expected Charge'] = (merged_data['Weight Slabs'] * merged_data['Forward Fixed Charge']) + (merged_data['Weight Slabs'] * merged_data['Forward Additional Weight Slab Charge']) + (merged_data['RTO Additional Weight Slab Charge']) + merged_data['RTO Fixed Charge']
```

```python
# Calculate charges billed by Courier Company
# ... (calculate forward charges, COD charges, total billed charges)
merged_data['Charges Billed'] = merged_data['Billing Amount (Rs.)']
```

```python
# Calculate difference between expected and billed charges
merged_data['Difference'] =  merged_data['Charges Billed'] - merged_data['Expected Charge']
```

```python
# Create resultant dataset
resultant_data = merged_data[['Order ID', 'AWB Code', 'Total Weight', 'Weight Slabs', 'Total Weight (Courier Company)', 'Weight Slabs (Courier Company)', 'Zone_x', 'Zone_y', 'Expected Charge', 'Charges Billed', 'Difference']]
```

```python
# Export resultant dataset
resultant_data.to_csv('output_data.csv', index=False)
```

```python
# Calculate total invoice amount for each category
correctly_charged_total = merged_data[merged_data['Expected Charge'] == merged_data['Charges Billed']]['Charges Billed'].sum()
overcharged_total = merged_data[merged_data['Expected Charge'] < merged_data['Charges Billed']]['Charges Billed'].sum() - correctly_charged_total
undercharged_total = merged_data[merged_data['Expected Charge'] > merged_data['Charges Billed']]['Charges Billed'].sum() - correctly_charged_total

# Calculate the count for each category
correctly_charged_count = merged_data[merged_data['Expected Charge'] == merged_data['Charges Billed']]['Order ID'].count()
overcharged_count = merged_data[merged_data['Expected Charge'] < merged_data['Charges Billed']]['Order ID'].count()
undercharged_count = merged_data[merged_data['Expected Charge'] > merged_data['Charges Billed']]['Order ID'].count()

# Create the summary table
summary_table = pd.DataFrame({
    'Category': ['Correctly Charged', 'Overcharged', 'Undercharged'],
    'Count': [correctly_charged_count, overcharged_count, undercharged_count],
    'Amount (Rs.)': [correctly_charged_total, overcharged_total, undercharged_total]
})

# Print the summary table
print(summary_table)
```
